```python
def odds(x):

    if x % 2:
        return True
    else:
        return False
```

```python
odds(7)
```

```python
odds(8)
```

```python
list(filter(odds, [1,2,3,4,5,6,7,8,9]))
```

```python
def çıkar(x):
    return x -= 5 # sadece koşul sonucu True ve Fasle olursa filter çalışır. Bu durumda hata verir
```

```python
çıkar(15)
```

```python
list(filter(çıkar, [1,2,3,4,5,6,7,8,9]))
```
